Here I'll convert the base 7 number 24036 into the normal base 10. Since we're not going to be doing any actual "calculations" on the number, it has been input as a string so we can pull out all the different digits to do separate calculations

```python
base10 = 0
digit = 0
b7num = '24036'

## The len function returns the length of the input string.
numLength = len(b7num)

for i in map(int, b7num):
  digit += 1

  base10 += i * (7**(numLength - digit))
base10
```

```python
base10 = 0
digit = 0
b16num = '72e3'

## The len function returns the length of the input string.
numLength = len(b16num)

for i in b16num:
    
    digit += 1
## This portion converts all the non-number characters to integers
    if i == 'a':
      convertedInt = 10

    elif i == 'b':
      convertedInt = 11

    elif i == 'c':
      convertedInt = 12

    elif i == 'd':
      convertedInt = 13

    elif i == 'e':
      convertedInt = 14

    elif i == 'f':
      convertedInt = 15

## Or if it is between 0-9 it will just return its value
    else:
      convertedInt = int(i)

    base10 += convertedInt * (16**(numLength - digit))
base10
```

Now, using two's complement, we will calculate 25 - 37. Of course in base 10 thats pretty easy to calculate: -12. But, we'll dive into two's complement

This portion of code would unfortunately need some slight modifications to work with different binary sequences

```python
binary1 = '00011001'      ## The binary equivalent of 25
binary2 = '00100101'      ## The binary equivalent of 37
binary2Complement = '1'    ## Declaration of 37's compliment variable. Note that it is declared as a string variable
digit = 0

for i in binary2:
  if (digit != 0):    ## This ensures we skip the first (farthest-left) digit in the string
    if (digit != 7):      ## This checks to see if we have reached the last digit
      if (binary2[digit] == '0'):     ## Checks to see if the selected digit is 0
        binary2Complement += '1'      ## If yes, appends a '1' to the output

      else:           ## If the selected digit is a 1, appends a '0' to the output
        binary2Complement += '0'
    else:             ## If we have reached the final digit, this will add a 1.
        binary2Complement += '1'    
  digit += 1
binary2Complement
```

Now that we've got our ducks in a row in that regard, we will work on the actual binary arithmetic that needs to take place.

I use an insane amount of nested if-else statements to handle different situations and accounting for a carry variable. I ran my code from right-to-left due to the need for carries. I then reversed the order to get the true answer.

```python
carry = '0'
invert = ''

for i in range(7, -1, -1):
  if carry == '0':    ## Doesn't contain a carry
    if binary1[i] == '0':
      if binary2Complement[i] == '0':
          invert += '0'
          carry = '0'
      else:
          invert += '1'
          carry = '0'
    else:
      if binary2Complement[i] == '0':
          invert += '1'
          carry = '0'
      else:
          invert += '0'
          carry = '1'
  else:                   ## Contains a carry
    if binary1[i] == '0':
      if binary2Complement[i] == '0':
          invert += '1'
          carry = '0'
      else:
          invert += '0'
          carry = '1'
    else:
      if binary2Complement[i] == '0':
          invert += '0'
          carry = '1'
      else:
          invert += '1'
          carry = '1'

result = invert[::-1]     ## The results are initially inverted: I reversed the order of the string to get the correct output
result
```
