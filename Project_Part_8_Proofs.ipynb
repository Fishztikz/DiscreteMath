We need to prove that (0, -4) is the *y*-intercept of the given function.

For starters, any and all *y*-intercepts MUST have an *x*-value of zero. For if the *x*-value were of a different value, it would not be the intersection of the *y*-axis as the *y*-axis always falls on *x* = 0. Therefore the *x*-coordinate of the point (0, -4) satisfies this particular condition.

One way we can go about proving this equation is plugging different *x*-values into this equation to get their respective *y*-values. For now, lets plug in *x*-values -3, -2, -1, 0, 1, 2, and 3.

```
def solveLinear(xVal):
  return (2*xVal) - 4

for i in range (-3, 4):
  print('(', i, ',', solveLinear(i),')', sep = '')
```

The code above iterates through the desired *x*-values to find their corresponding *y*-values. The for loop calls the *solveLinear* function for each value of *x*, and the print statement calculates the *y*-value by plugging in the given *x* and prints it.

We can see that this function actually returns the expected *y*-intercept of (0, -4), but thats not truly "proving" anything. We can also have python generate a graph.


```
import numpy as np
import matplotlib.pyplot as plt

def solveLinear(xVal):
  return (2*xVal) - 4

x = np.arange(-3,4,1)
plt.plot(x,solveLinear(x), '--bo')
```

While it looks a touch different than the Desmos graph earlier, the values are all the same. As can be clearly seen, this is a linear graph. Further evidence that this is a linear graph can be found by the actual equation:
$$y = 2x - 4$$

Since neither *x* nor *y* have exponents or other modifiers, this is a linear equation and graphically can only produce a straight line. Therefore, if we find the midpoint between *x*-value -1 and 1, we would get the *y*-intercept. We can use the midpoint formula:
$$(x_m, y_m) = (\frac{x_1 + x_2}{2},\frac{y_1 + y_2}{2})$$

$$(x_m, y_m) = (\frac{-1 + 1}{2},\frac{-6 - 2}{2})$$

$$(x_m, y_m) = (\frac{0}{2},\frac{-8}{2})$$

$$(x_m, y_m) = (0,-4)$$

The midpoint between the two values surrounding *y* = 0 is (0, -4). As we saw earlier we can also plug 0 into the function for *x* and will get zero to get another method of proof.

$$y = 2(0) - 4$$
$$y = -4$$

Therefore, we've proved that the *y*-intercept of y = 2x - 4 is in fact -4. Q.E.D
