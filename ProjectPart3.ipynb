For this project, I'll be using the 6 different-colored highlighters I keep in my college bag. The colors I have are yellow, blue, pink, orange, green, and purple. I'll be using my two favorite, basically-dried-out highlighters for the majority of my project. The probability of pulling the yellow highlighter out of my bag is 1/6, and 1/6 for the orange highlighter as well. The chance for neither a yellow or orange is 4/6, or 2/3.

```python
 prob =[1/6,1/6,2/3]

 sum(prob)
```

What's the probability that I pull out either the yellow or orange highlighter?

```python
ProbabilityR = 0

for i in [0,1]:
  ProbabilityR = ProbabilityR + prob[i]

ProbabilityR
```

What about the intersections and unions of yellow or orange highlighters? Since I only have 1 of each color, lets just assume that I put the highlighter I pull back into the bag after each trial.

Here's the likelihood of getting the yellow highlighter, and subsequently getting the orange highlighter, or the yellow highlighter again.

$$P(O\cap R)$$

```python
intersection = prob[0]*ProbabilityR
intersection
```

What would the probability be if I pulled out the yellow highlighter on the first draw, and would pull out a yellow or orange highlighter on the second trial?
$$
P(O|Y)=\frac{P(O\cap Y)}{P(Y)}
$$

```python
YGO = intersection/prob[0]

YGO
```

How about the probability of getting a yellow highlighter, given either the yellow highlighter or orange highlighter was pulled out first?

```python
OGY = intersection/ProbabilityR

OGY
```

Now we'll take a look at Bayes' formula. The code below compares the results we received from the previous function, and compares it to Bayes' formula. It should return true if they are equal.

```python
OGY ==(prob[0]* YGO)/ProbabilityR
```

How about the complement of getting the yellow highlighter? With the shape my yellow highlighter is in (dry) it'd probably be a good idea to use different colors anyways. Complement can be found relatively easily-- just subtract the probability of pulling the yellow highlighter out by one.

```python
complement = 1 - prob[0]
complement
```

Now for some theoretical simulations, lets see what would happen if I repeated the trial 10 times. Remember that I'm adding the highlighters back into the bag after each trial. Therefore, each of the 10 trials should have identical chances.

I will start by getting the frequency counts. This is found by multiplying the probability of each outcome by the number of outcomes. In my case, there are 6 outcomes.

```python
pCount = [6*i for i in prob]
pCount
```

```python
from random import choices

highlighters =  choices(['Yellow','Orange','Other'],pCount,k=10)
print(highlighters)

cYellow = 0
cOrange = 0
cOther = 0

for i in highlighters:
  if i =='Yellow':
    cYellow += 1
  elif i == 'Orange':
    cOrange += 1
  elif i == 'Other':
    cOther += 1

estProbability = [cYellow/len(highlighters),cOrange/len(highlighters),cOther/len(highlighters)]
estProbability
```

The results are different each time, but after a few different runs it seems the chances we calculated are pretty close to accurate.
