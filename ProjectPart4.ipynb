Here we find the factorial of 10-- how many different sequences can I choose all 10 items? 

```python
import math

math.factorial(10)
```

How many different outcomes could we have if order mattered, but we replaced the item after each pick? To get that answer, we would just multiply the amount of choices for the first pick (10) and the amount of choices for the second pick (also 10, since we replaced the first pick). We would get 
$$10 * 10 = 10^2$$

```python
10**2
```

What if order didn't matter but we stopped replacing items? This is known as a combination and we can use 10 choose 2, which is the amount of ways we can choose 2 items out of a set of 10.
$$\frac{x!}{(n-k)!k!}$$

```python
math.factorial(10) / ((math.factorial(10 - 2)) * math.factorial(2))
```

Flipping around, what if order DID matter and we didn't replace any items? This is called a partial permutation.
$$\frac{x!}{(n-k)!}$$

```python
math.factorial(10)/(math.factorial(10 - 2))
```

In esssense, a combination refers to the scenario in which order doesn't matter. Permutations however, involve scenarios in which order DOES matter.

$$(x^5)(-y^4)$$
