For this project, I'll be using the current predominant V8 engine's from the Big Three (Ford, Chevrolet, and Chrysler) and comparing the base architectures of their engines


*   (P) Ford - 5.0L Coyote, Dual-Overhead Cam (DOHC) V8
*   (Q) Chevy - 6.2L LT1, Cam-in-Block Pushrod V8
*   (R) Chrysler - 6.4L Hemi, Cam-in-Block Pushrod V8


Some of these terms are probably unfamilar to most, but in essense it is the placement of the camshaft(s) within an engine. You don't need to know the "what" or "why" though, just know that Chevy and Chrysler both use the same Cam-in-Block architecture while Ford does not.

Going off of that, I'll set up truths from the statement "Engines that utilize Cam-in-Block architecture."


```python
Coyote = False
LT1 = True
Hemi = True
```

We'll start off with  **R∧P**, which can be stated as "Chrysler's 6.4L Hemi and Ford's 5.0L Coyote utilize Cam-in-Block architecture."

```python
Hemi and Coyote
```

Now we can do **R∨¬Q**, or "Chrysler's Hemi or NOT Chevy's LT1 utilize a Cam-in-Block architecture. Quite a confusing sentence to read for sure.

```python
Hemi or not(LT1)
```

not(LT1) returns false, but since Hemi returns true, this truth statement returns true.

Now we can find **R⇒P**, which means "If Chrysler's Hemi is a Cam-in-Block engine, Ford's Coyote must be a Cam-in-Block as well." We can substitute **R⇒P** as **¬R∨P** since these are equivalent statements.

```python
not(Hemi) or Coyote
```

Now to build a few truth tables.

Now that we got everything imported, here's a truth table for  **¬(r⊕p)⇒(p∨q)**.

NOT (ONLY Chrysler's Hemi or Ford's Coyote is a cam-in-block engine) implies that either Ford's Coyote or Chevy's LT1 is a cam-in-block engine. I'm not the best at putting truth statements into words as you can tell, but the truth table is below.

```python
truthtable = ttg.Truths(['p', 'q', 'r'], ['r xor p', '~(r xor p)', 'p or q', '(~(r xor p)) => (p or q)'], ints = False)
truthtable.as_pandas().style.set_properties(**{'text-align': 'center'}).hide_index()
```

As can be seen, for our example (P = False, Q = True, and R = True) the result of this truth statement returns True.
