### Part E

This bit of code will determine if every model in the Ford list is in the UniversalSet. If a model is in the Ford set but not in the Universal set, Ford would not be a subset of the Universal set (it would return false).

```
def IsASubset(A, B):
  for model in A:
    if model not in B:
      return False
  return True

IsASubset(Ford, UniversalSet)
```

```
True
```

Yes, subsets do exist. The other two brand sets, Dodge and Chevrolet, are subsets of the universal set as well.
