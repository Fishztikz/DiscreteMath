A universal set UniversalSet has been created. It includes all models in the three sets, and additional performance models from other brands, such as Buick, Mercury, Chrysler, Jeep, and Toyota.

Intersections and Unions are created between the three sets. First is A∪B, which will create a set of all models listed in the Chevrolet and Ford sets. This will append each model from both Chevrolet and Ford sets into a new list called AUB. It is then converted into a set.



```python
AUB = []
for model in Chevrolet:
  AUB.append(model) # Add all performance models from set Chevrolet
for model in Ford:
  AUB.append(model) # Add all performance models from Ford
  
AUB = set(AUB) # Create a set from list.
AUB 
```

```python
AUC = []
for model in Chevrolet:
  AUC.append(model) # Add all performance models from set Chevrolet
for model in Dodge:
  AUC.append(model) # Add all performance models from Dodge

AUC = set(AUC) # Create a set from list.
AUC 
```

### Part E

This bit of code will determine if every model in the Ford list is in the UniversalSet. If a model is in the Ford set but not in the Universal set, Ford would not be a subset of the Universal set (it would return false).

```python
def IsASubset(A, B):
  for model in A:
    if model not in B:
      return False
  return True

IsASubset(Ford, UniversalSet)
```

Yes, subsets do exist. The other two brand sets, Dodge and Chevrolet, are subsets of the universal set as well.

### Part F

Now lets create a powerset. We'll make a powerset of Chevrolet. A powerset creates sets of every possible combination of models within the set. This uses two a for loop nested within another for loop to each run through the different models. The results are lengthy.

```python
powerset = [[]]
for model in Chevrolet:
    for eachSet in powerset:
      powerset = powerset + [eachSet + [model]]
print(powerset)
```

The amount of sets that would be created by a superset can be calculated by 2^n, where n is the number of items that are included in the set. If a powerset of the Universal set were to be created, it would generate 2^25 different sets, which comes out to 33554432 sets. That's quite a handful.

### Part A
For this example, I will be using the function f(x) = (1/4)x * sin(x) + 1. This function looks a little crazy on the graph, looks like a radiowave-esque shape. More on that in a little bit.

```python
import math
import numpy as np

def reverb(x):
  return ((1/4) * x * np.sin(x) + 1)

reverb(0)
```

With this code, the function was defined and calculated at x = 0. The output was just a simple 1. Pretty lame.

```python
import matplotlib.pyplot as plt

x = np.arange(-4,4,.1)

plt.plot(x,reverb(x))
```

The np.arrange function provides a graph, where the first two inputs are the range of the graph that is shown (in this case, -4 and 4). The final input essentially describes the detail of the graph. The 0.1 means my reverb function will calculate the x-value on every tenth between -4 and 4, and graph those points. If I were to set that to just 1, it will only calculate on the whole numbers (and thus would be a much less detailed graph).

This function is not bijective, as multiple x-values can share a y-value

## 3 - Dictionaries
Now we can work on dictionaries. Inspired by my lists from the Sets project, I will be using some of the same muscle/pony cars from those lists. But this time I will be pairing the cars with their respective engines

```python
A = ['Mustang GT', 'Challenger R/T', 'Camaro ZL1', 'Mustang GT350', 'Challenger Demon', 'Corvette']
B = ['5.0L Coyote', '5.7L Hemi', '6.2L LT4', '5.2L Voodoo', '6.2L Hellcat', '5.5L LT6']

dictionary = dict({a:b for a,b in zip(A, B)})

dictionary
```

In this example, the dictionary takes the car's name as the key, and appends a value to each. In this case, the value is their respective motors. 

The method used here is just taking two sets and essentially "combining" them. You could also just create a dictionary outright by just putting the values in manually, but where's the fun in doing it the easy way? Here's a similar dictionary, but this time the values have been added manually. Note that I have removed the trim levels from the model's name in this dictionary to match the models defined in the Sets project. You'll see why in a bit.

```python
outrightDictionary = dict({'Mustang':'5.2L Voodoo','Challenger':'6.2L Hellcat','Camaro':'6.2L LT4', 'Corvette':'5.5L LT6'})

outrightDictionary
```

Below, we create a new set that is identical to the Chevrolet set in the first project, but the dictionary will act on this set--It replaces the name of the model with it's respectively-defined engine in the dictionary. As you can see Camaro and Corvette have been changed. This example is not reflexive or transitive.

```python
newChevroletList = [outrightDictionary.get(x,x) for x in Chevrolet ]
newChevroletList
```

Now we'll give Dodge the same treatment. If it runs as expected only "Challenger" will be replaced.

```python
newDodgeList = [outrightDictionary.get(x,x) for x in Dodge ]
newDodgeList
```

As you can see, both sets have been changed. I didn't change everything in the lists however so neither of these freshly-minted lists are onto. Onto in this example would require each VALUE to be mapped to a certain KEY, and neither of these lists map all values in the dictionary. As such, neither are bi-jective (one-to-one and onto). However, if the UniversalSet from the Sets project was used, that list would in fact be onto, as every element from the dictionary would be used in the relation.
