For this project, I will be graphically showing the relationships of all the engine choices for each generation of Ram trucks. I've been debating an engine swap for my 2006 Ram, so a graphical display of this subject is definitely helpful in showing the difficulty of each.

Instead of identifying the individual model years of the trucks, I will be dividing them into their respective "Generations," which is a span of about 5-10 years in which little fundamental changes are made to a vehicle. The generations go as follows:

* 1st Gen: 1981-1993
* 2nd Gen: 1994-2001
* 3rd Gen: 2002-2008
* 4th Gen: 2009-2018
* 5th Gen: 2019-2022




```python
import graphviz

eng = graphviz.Digraph('Engines')

# Here I just listed the different generations of trucks.
eng.node('A', '1st Gen')
eng.node('B', '2nd Gen')
eng.node('C', '3rd Gen')
eng.node('D', '4th Gen')
eng.node('E', '5th Gen')

# and here are their respective engines.
eng.node('F', '318 LA')
eng.node('G', '360 LA')
eng.node('H', '3.9L Magnum')
eng.node('I', '318 Magnum')
eng.node('J', '360 Magnum')
eng.node('K', '8.0L Magnum')
eng.node('L', '3.7L Magnum')
eng.node('M', '4.7L Magnum')
eng.node('N', '5.7L Hemi')
eng.node('O', '8.3L Viper V10')
eng.node('P', '5.9L Cummins I6')
eng.node('Q', '6.4L Hemi')
eng.node('R', '3.6L Pentastar')
eng.node('S', '3.0L EcoDiesel')
eng.node('T', '6.7L Cummins I6')


# Now we will create the relationships
eng.edges(['AF', 'AG', 'AP', 'BH', 'BI', 'BJ', 'BK', 'BP', 'CJ', 'CL', 'CM', 'CN', 'CO', 'CP', 'CT', 'DL', 'DM', 'DN', 'DQ', 'DR', 'DS', 'DT', 'EN', 'ER', 'EQ', 'ES', 'ET'])
eng
```

Thats a lot to unpack. I'll describe the path that my truck takes as an example. My truck is a 2006 model year, meaning it falls in the 3rd gen category (2002-2008, referred to as node ***C*** in the code). During that time there were 7 different engine options (Realistically just three, but I don't want to get overly complicated here). My truck came optioned with the 3.7L Magnum (node ***L***), the smallest engine of the bunch. So according to the code, my trucks path would be ***CL***.


As mentioned earlier, the 3rd gen category had 7 engine options over its 6-year lifespan. This can be easily seen on the graph, as there are 7 engines listed that have a direct relationship with the 3rd gen node, and 7 lines coming out of the node. Therefore, the degree of vertex ***C*** (3rd gen) is 7.


While the code automatically added arrows to the visualization, this is best read as an undirected graph. Therefore, a couple of cycles could technically be made out of this graph. For example, the 2nd gen and 3rd gen share two common engines-- The 360 Magnum and the 5.9L Cummins I6. The cycle could go as follows:

2nd Gen -> 360 Magnum -> 3rd Gen -> Cummins 5.9L I6 -> 2nd Gen, or:
BJ, JC, JP, PB

This does make some sense, as quite a lot of hardware (and software) is shared between the two generations due to these two engines. Technically, a cycle could be made from the 5th Gen category all the way to the 2nd Gen category and back, as each of those truck generations share at least two engines with the generation before. 

To get even more technical, the size of the engine bay of the 2022 Ram (a 5th gen) can be almost directly attributed to the size of the engine bay in the 1981 Ram. Which we could go even further and say the Dodge trucks of the 1950s and 1960s influenced the size of the engines used in the 1st gen Rams, which by extension influences the size of the engines used in the trucks today.
